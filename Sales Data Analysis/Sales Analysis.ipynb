# Project: Analysis and Visualization of 12 Months of Sales 

## Learning Objective
Using Pandas and Matplotlib to clean, analyze and answer business questions of 12 months worth of sales data. The dataset
contains 12 different files with thousands of electronic purchases broken down into month, product type, cost, purchase
address, etc

### Import Necessary Libraries

```python
import pandas as pd
import os
```

### Merging 12 months of sales into 1 single file

```python
# Inspect one file fromt the directory
pd.read_csv('./Data/Sales_April_2019.csv')
```

```python
pd.read_csv('./Data/Sales_May_2019.csv')
```

It looks like all the CSV files in our directory have the same columns and dimension! Therefore, we can use pd.concatenate to combine all 12 files into 1 CSV file.

```python
# Create a new DataFrame
df_merge = pd.DataFrame()

# Concatenate 12 CSV files into 1 file
files = [file for file in os.listdir('./Data')]

for file in files:
    df = pd.read_csv('./Data/'+file)  # Read all the files in our directory
    df_merge = pd.concat([df_merge, df])   # Merge all the files into a new Dataframe
    
# Inspect the new merge DataFrame
df_merge
```

We can also export the new merge file into csv for further analysis as follow:

```python
df_merge.to_csv('Combined Data.csv', index=False)
```

### Cleaning up the updated DataFrame

```python
combined_data = pd.read_csv('Combined Data.csv')
combined_data.head()
```

```python
# Check for NaN values
combined_data.isna().sum()
```

```python
# Drop rows of Nan values
combined_data = combined_data.dropna(how='all')
```

```python
# Select rows that doesn't have the valid Order Date
combined_data[combined_data['Order Date'].str[0:2] == 'Or']
```

```python
# Get rid of the rows with invalid Order Date
combined_data = combined_data[combined_data['Order Date'].str[0:2] != 'Or']
```

### Let's try to see what the best month for sales was, and how much revenue was earned in that month.

Since our combined file doesn't have a colum for month, we would create a new column that store the month from all the
transactions recorded.

```python
combined_data['Month'] = combined_data['Order Date'].str[0:2].astype('int32')
combined_data.head()
```

To see what is the total sales for each month, let's add a new column called Sales.

```python
# Convert Quantity Ordered into int
combined_data['Quantity Ordered'] = pd.to_numeric(combined_data['Quantity Ordered'])

# Convert Price Each into int
combined_data['Price Each'] = pd.to_numeric(combined_data['Price Each'])

# Check the type results
combined_data.info()
```

```python
# Add a new column Sales
combined_data['Sales'] = combined_data['Quantity Ordered'] * combined_data['Price Each']
combined_data.head()
```

```python
# Find the total sales per month
monthly_sales = combined_data.groupby('Month').sum().reset_index()

# Find the month with the highest sales
monthly_sales.sort_values(by='Sales', ascending=False)
```

I will try to visualize our results with a bar chart for better comparison.

```python
# Import Matplotlib for visualization
import matplotlib.pyplot as plt

x = monthly_sales['Month']
y = monthly_sales['Sales']

plt.figure(figsize=(17,8))
plt.xticks(monthly_sales['Month'])
plt.ylabel('Sales in USD ($)', fontsize=16, weight='bold')
plt.xlabel('Month', fontsize=16, weight='bold')
plt.title('Monthly Sales Comparison', fontsize=24, weight='bold')


plt.bar(x, y, color='cornflowerblue')
plt.show()
```

### Insight: 
It appears that December has the highest volume of sales across the years. This really makes sense since end of year
usually has a lot more promotions and higher demands for new consumptions of electronic products (i.e: Christmas gifts, New Year Celebration, etc)

### We can also look into what location has the highest sales revenue in the US!

In order to create a City column, we can use pd.str.split() to extract the second values from the column of Purchase
Address. "Expand" argument is equal to True since we want the DataFrame to expand its dimensionality (i.e: creating a 
new column).                                                                                                   

```python
# Create a new column with Location using pd.str.split()
new = combined_data['Purchase Address'].str.split(',', expand=True)

# Include an acronym to avoid city duplicates
combined_data['City'] = new[1] + ' ' + new[2].str.split(' ', expand=True)[1]
combined_data
```

```python
# Find total sales according to location
city_sales = combined_data.groupby('City').sum().reset_index()

# Find the city with the highest sales
city_sales = city_sales.sort_values(by='Sales', ascending=False)

city_sales
```

```python
import matplotlib.pyplot as plt

x = city_sales['City']
y = city_sales['Sales']

plt.figure(figsize=(17,8))
plt.ylabel('Sales in USD ($)', fontsize=16, weight='bold')
plt.xticks(rotation=45, fontsize=13)
plt.xlabel('Location', fontsize=16, weight='bold')
plt.title('City Sales Comparison', fontsize=24, weight='bold')

bar_chart = plt.bar(x, y, color='darkgreen')
bar_chart[0].set_color('darkred')
plt.show()
```

### Insight:

San Francisco CA seems like the hot spot for electronics products. I speculate that this might be due to increasing demands for tech products from
tech companies and their employees in the Bay Area and San Francisco Sillicon Valley.

## I'm also curious about the time of purchase order and how we can maximize store revenue based on the displaying product time. We can investigate what time period we should display advertisements to increase likelihood of customers' purchasing our electronics products.

```python
# Conver the Order Date column into a datetime type
combined_data['Order Date'] = pd.to_datetime(combined_data['Order Date'])

# Double check our results
combined_data.info()
```

```python
# Create a 'Hour' column from the 'Order Date'
combined_data['Hour of Purchase Order'] = combined_data['Order Date'].dt.hour

# Create a new column 'Count' so that we can later aggregate the time for order purchase
combined_data['Count'] = 1
combined_data.head()
```

```python
# Create a 'Hour' key for our graph using a list comprehension
hours = [hour for hour, df in combined_data.groupby('Hour of Purchase Order')]
hours
```

```python
# Create a line chart
import matplotlib.pyplot as plt

plt.figure(figsize=(17,8))
plt.plot(hours, combined_data.groupby('Hour of Purchase Order').count())

plt.ylabel('Number of Orders', fontsize=16, weight='bold')
plt.xlabel('Hour', fontsize=16, weight='bold')
plt.title('Distribution of Hour Based on the Number of Purchases', fontsize=24, weight='bold')
plt.xticks(hours)
plt.grid()
plt.show()
```

### Insight:

Our electronics store usually sells the most products around 12pm and 7pm. This might be attributed to the daily breaks
that people usually have during a normal working day (lunchtime and after work). Hence, in order to increase customers'
likelihood to purchase a product, we can place more ads around 11am-12pm and 6pm-7pm.

## Looking at the Product column, which products are most often sold together?

To see what are the products that are sold together, we can check the duplicates of the 'Order ID' column. 'False' is 
parsed to 'keep' since we want to see and keep all duplicates.

```python
# Create a new DataFrame that contains duplicates of Order ID
df_order = combined_data[combined_data['Order ID'].duplicated(keep=False)]

df_order.head()
```

```python
# Create a new column that combine all products that are sold together in one order
df_order['Product Combination'] = df_order.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

df_order.head(10)
```

```python
# Drop duplicates in the Order ID now that we have the Product Combination
df_order = df_order.drop_duplicates(subset=['Order ID'])

df_order.head()
```

```python
# Count the unique pairs of products into a Python dictionary
from itertools import combinations
from collections import Counter

count = Counter()

for row in df_order['Product Combination']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))   # Get the combination for 2 products

# Get the 10 most common combinations of 2 products
for key, value in count.most_common(10):
    print(key, value)
```

### Insight:

Iphones and Lighting Charging Cable are most often purchased together in our stores. Hence, we can attempt to boost
sales by create attractive bundle promotion to trigger more purchases for these two products.

## Then which product is our best-seller? What can be the reasons behind its trend?

```python
# Define Products and their Quantity
product_name = combined_data.groupby('Product')
product_quantity = combined_data.groupby('Product')['Quantity Ordered'].sum()

# Create a list comprehension of products
product = [product for product, df in product_name]

plt.figure(figsize=(17,8))
plt.bar(product, product_quantity, color='cornflowerblue')
plt.ylabel('Number of Orders', fontsize=16, weight='bold')
plt.xlabel('Product', fontsize=16, weight='bold')
plt.title('Best Selling Products', fontsize=24, weight='bold')
plt.xticks(rotation=70, fontsize=13)
plt.show()
```

Looking at the bar chart above, I want to also compare the electronics price on top of their purchase orders. Let's see
how we can do that!

```python

x = [product for product, df in product_name]
y = combined_data.groupby('Product')['Quantity Ordered'].sum()

plt.figure(figsize=(17,8))
plt.bar(x, y, color='cornflowerblue')
plt.title('Best Selling Products', fontsize=24, weight='bold')
plt.xlabel('Product', fontsize=16, weight='bold')
plt.xticks(rotation=70, fontsize=13)
plt.ylabel('Number of Orders', fontsize=16, weight='bold')

x = [product for product, df in product_name]
y1 = combined_data.groupby('Product')['Price Each'].mean()
axes2 = plt.twinx()
axes2.plot(x, y1, color='darkred')
axes2.set_ylabel('Price Each USD ($)', fontsize=14, weight='bold')

plt.show()
```

### Insight:

It is evident from our graph that the number of orders and price of electronics products demonstrate a negative correlation.
The higher the price, the lesser number of orders that our customers would purchase and vice versa!
