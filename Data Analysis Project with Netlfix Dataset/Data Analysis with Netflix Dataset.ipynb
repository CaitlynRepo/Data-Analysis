## Insight:

It seems that the total number of movies and TV shows added is increasing every year on Netflix platform. Noted that
our dataset hasn't updated total number for 2021, and hence, demonstrates a low value for 2021. One shortcoming of this graph is that it doesn't clearly demonstrate exactly how many Movies and TV Shows within the total count for each year. Let's see how we can make a clearer chart.

```python
# Add a column 'Release Year' to the original Netflix DataFrame
df['Release Year'] = df['Release_Date'].dt.year
df.head()
```

```python
# Create a DataFrame of Movie only
netflix_movies_df = df[df['Category'].str.contains("Movie")]

# Create a DataFrame that counts total movies released each year
movies_year_df = netflix_movies_df['Release Year'].value_counts().to_frame().reset_index().rename(columns={"index": "Year", "Release Year":"Movie Count"})
movies_year_df
```

```python
# Create a DataFrame of TV Shows only
netflix_shows_df = df[df['Category'].str.contains("TV Show")]

# Create a DataFrame that counts total TV shows released each year
shows_year_df = netflix_shows_df['Release Year'].value_counts().to_frame().reset_index().rename(columns={"index": "Year", "Release Year":"TV Show Count"})
shows_year_df
```

```python
# Merge two DataFrame so that we have both Movies and TV Show Counts
movies_tv_shows = pd.merge(shows_year_df, movies_year_df, on='Year', how='outer')

#df.sort_values(['Total Due'], inplace=True)
movies_tv_shows = movies_tv_shows.sort_values(['Year'])
movies_tv_shows
```

```python
# Plot a stacked bar chart, with specific figsize
movies_tv_shows.set_index('Year').plot(kind='bar', stacked=True, figsize=(17,8))
plt.title('Number of Movies and TV Shows that Are Released on Netflix Each Year', fontsize=24, weight='bold')
plt.xlabel('Year', fontsize=16, weight='bold')
plt.ylabel('Count of Shows', fontsize=16, weight='bold')
plt.show()
```

## Insight:
Now we can see a bit clearer that while the total number of content added has increased significantly since 2015, the growth
for Movie content is much higher than that of TV Shows. It looks like Netflix has strategically focused on Movies rather than
television shows in the recent years.

What if I want to count top 10 directors individually? Meaning that Raul Campos and Jan Suter would be seperated and
counted based on the number of content ('Title') they worked on individually.

```python
# Get the top 10 directors based on the number of titles they have worked on 
filtered_directors.value_counts().nlargest(10)
```

```python
# Create a countplot in Seaborn to illustrate our result
plt.figure(figsize=(15,5))
sns.countplot(y=filtered_directors, order=filtered_directors.value_counts().nlargest(10).index, palette='Set2')
plt.title('Top 10 Directors That Produced The Most Content On Netflix', fontsize=20, weight='bold')
plt.ylabel('Director', fontsize=16, weight='bold')
plt.xlabel('Count of Shows', fontsize=16, weight='bold')
plt.show()
```

## Insight:
The countplot is a bit clearer than the previous bar chart since it distinctly indicates individuals who produced the
most shows on Netflix platform. Jan Suter is the directors with the most movies produced, followed by Raul Campos and Marcus Raboy
with 19 and 16 shows respectively.

## Note:
I called 'order' argument in the sns.countplot to plot the 'Director' names since these are categorical values. 'Order'
is used for list of string (in this case the Director names). Specifying the list of strings as x or y would yield errors :D

## 9. What Are The Different Ratings Defined by Netflix?

```python
# Show all the unique values in the Rating column
df['Rating'].unique()
```

```python
# Show total number of content with all types of ratings
df.groupby('Rating')['Title'].count().sort_values(ascending=False)
```

```python
# Create a DataFrame for all ratings, categorized with Movie and Tv shows
rating_movies_tv_shows = df.groupby('Rating')['Category'].apply(lambda x: x.value_counts()).to_frame().reset_index().rename(columns={"index": "Rating", "Category":"Count of Shows", "level_1":"Category"})
rating_movies_tv_shows
```

```python
# Transpose the Category columns into two 'Movie' and 'TV Show' columns
rating_movies_tv_shows = rating_movies_tv_shows.pivot_table(index='Rating', columns='Category', values='Count of Shows').reset_index()
rating_movies_tv_shows
```

```python
rating_movies_tv_shows.set_index('Rating').plot(kind='bar', stacked=True, figsize=(17,10))
plt.title('Number of Movies and TV Shows Categorized by Ratings', fontsize=24, weight='bold')
plt.xlabel('Ratings', fontsize=16, weight='bold')
plt.ylabel('Count of Shows', fontsize=16, weight='bold')
plt.show()
```

## Insight:
TV-MA (TV - Mature Audience: unsuitable for children under 17) comprises of the largest portion of Netflix content. Movie
spreads for almost all types of ratings. ranging from G to TV-17 or TV-MA. Meanwhile, TV shows predominantly focus on TV-14,
TV-MA and TV-PG content.

## 10. What Is The Longest Duration Of A Movie/Show On Netflix?

```python
# Check unique values on the column 'Duration'
df['Duration'].unique()
```

Since we have different units of measurements here (min and Seasons), I would attempt to create new columns for each 
measurements and get the max values of each. Let's so how we can do that.

```python
df[['Unit', 'Measurement Type']] = df['Duration'].str.split(' ', expand=True)
df.head()
```

```python
# Convert the 'Unit' column into numeric value
df['Unit'] = pd.to_numeric(df['Unit'])

# Find the longest duration for each type of measurements
df.groupby('Measurement Type').agg({'Unit':'max'}).reset_index()
```

```python
# Filter out show(s) with 16 Seasons
df.loc[(df['Unit'] == 16) & (df['Measurement Type'] == 'Seasons')]
```

```python
# Filter out show(s) that has the duration of 312 min
df.loc[(df['Unit'] == 312) & (df['Measurement Type'] == 'min')]
```

## Insight:
Based on our output, we can conclude that the longest running shows on Netflix are Grey's Anatomy with 16 seasons and Black Mirror
Bandersnatch whose duration is 312 min. One interesting fact about Black Mirror: Bandersnatch's duration (that I checked online) is that it has numerous permutations and alternate endings. Based on the options you take for the story, you 
can spend either 40 minutes or more than 2 hours (and hence, the 312 min) to complete the movie! Quite interesting huh? :D
